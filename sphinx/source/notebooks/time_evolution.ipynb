### Time evolution steps

#### Electron potential calculation step (1-D, 2 electrons)

When the dimension is 1, the distance between particles are calculated by a simple absolute function, rather than a norm.

```python
import crsq.firstq as firstq
t = firstq.FirstQIntegrator()
t.set_dimension(1)
t.set_coordinate_bits(3)
t.set_num_particles(2,1)
t.set_num_energy_configurations(4)
t.set_energy_configuration_weights([0.25, 0.25, 0.25, 0.25])
t.set_max_particle_charge(1)
t.set_num_atom_iterations(4)
t.set_num_elec_per_atom_iterations(16)
t.make_dummy_orbitals()
t.make_dummy_nuclei_data()
t.calculate_sizes()
t.allocate_registers()

reg_set = t._build_elec_potential_step()
reg_set.circuit.draw(scale=0.6)
```

#### Electron potential calculation step (2-D, 2 electrons)

When the dimension is 2 or 3, a norm must be used for distance calculation, which require squares and square roots.
Squares and square roots both require allocating long term work registers, that need to be occupied until the
inverse operation of the gate is applied.

```python
import crsq.firstq as firstq
t = firstq.FirstQIntegrator()
t.set_dimension(2)
t.set_coordinate_bits(3)
t.set_num_particles(2,1)
t.set_num_energy_configurations(4)
t.set_energy_configuration_weights([0.25, 0.25, 0.25, 0.25])
t.set_max_particle_charge(1)
t.set_num_atom_iterations(4)
t.set_num_elec_per_atom_iterations(16)
t.make_dummy_orbitals()
t.make_dummy_nuclei_data()
t.calculate_sizes()
t.allocate_registers()

reg_set = t._build_elec_potential_step()
reg_set.circuit.draw(scale=0.5)
```
